##Работа с пропусками

Заранее убедившись, что проруски в колонках price_usd и insurance_amount_usd обусловлены тем, что в price и insurance_amount уже указаны значения в usd, я просто заполняю пропуски в колонках price_usd и insurance_amount_usd теми же значениями что и в price и insurance_amount

То же самое делаю с колонкой loss_payout_amt_usd, а затем заполняю оставшиеся пропуски в loss_payout_amt_usd и loss_payout_am ноликами, т.к. убытка не было.

```python
df['price_usd'] = df['price_usd'].fillna(df['price'])
df['insurance_amount_usd'] = df['insurance_amount_usd'].fillna(df['insurance_amount'])
df['loss_payout_amt_usd'] = df['loss_payout_amt_usd'].fillna(df['loss_payout_amt'])
df['loss_payout_amt'] = df['loss_payout_amt'].fillna(0)
df['loss_payout_amt_usd'] = df['loss_payout_amt_usd'].fillna(0)
msno.bar(df)
```

##Кодировка категориальных признаков

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
df = df.drop(columns=['contract_num', 'client_id'])
```

```python
le = LabelEncoder()
pr_pred = le.fit_transform(df['product_name'])
df['product_name'] = pr_pred

le = LabelEncoder()
pr_pred = le.fit_transform(df['contract_status'])
df['contract_status'] = pr_pred

le = LabelEncoder()
pr_pred = le.fit_transform(df['currency_name'])
df['currency_name'] = pr_pred

le = LabelEncoder()
pr_pred = le.fit_transform(df['sex'])
df['sex'] = pr_pred

le = LabelEncoder()
pr_pred = le.fit_transform(df['country'])
df['country'] = pr_pred

df
```

##Работа с аномалиями

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.ensemble import IsolationForest
```

```python
lof = LocalOutlierFactor(n_neighbors=5)
pred = lof.fit_predict(df)
```

```python
df['anomaly'] = pred

df['anomaly'] = df['anomaly'].replace(1, 0)
df['anomaly'] = df['anomaly'].replace(-1, 1)

df
```
